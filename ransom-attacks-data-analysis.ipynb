# **Import libraries and data**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv(r"/kaggle/input/ransomware-attacks/Ransomware Attacks - Ransomware Attacks.csv")
df.describe(include="all")
df.info()
df.shape
df.head(5).T
df.isnull().sum()
```

# **Visualize data**

```python
sns.heatmap(df.isna())
```

```python
sector = df['sector'].value_counts()
sector
plt.figure(figsize=(20, 15))
plt.xticks(rotation=70, fontsize=10)
sns.set_palette("colorblind")
sns.barplot(x=sector.index, y=sector.values)
plt.xlabel('Sector')
plt.ylabel('Count')
plt.title('Count of each sector')

plt.show()
```
